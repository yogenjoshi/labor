```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style
plt.style.use('ggplot')
```

```python
laborRegion_df = pd.read_excel('LaborRegion.xlsx')
head = laborRegion_df.columns
print(head)
```

```python
riyadh_df = laborRegion_df[laborRegion_df.Region == 'Riyadh']
makkah_df = laborRegion_df[laborRegion_df.Region == 'Makkah'] 
madinah_df = laborRegion_df[laborRegion_df.Region == 'Madinah']   
qassim_df = laborRegion_df[laborRegion_df.Region == 'Qassim']
east_df = laborRegion_df[laborRegion_df.Region == 'EastProv']
asir_df = laborRegion_df[laborRegion_df.Region == 'Asir']
tabuk_df = laborRegion_df[laborRegion_df.Region == 'Tabuk']
hail_df = laborRegion_df[laborRegion_df.Region == 'Hail']
north_df = laborRegion_df[laborRegion_df.Region == 'NorthBord']
jazan_df = laborRegion_df[laborRegion_df.Region == 'Jazan']
najran_df = laborRegion_df[laborRegion_df.Region == 'Najran']
albaha_df = laborRegion_df[laborRegion_df.Region == 'AlBaha']
aljouf_df = laborRegion_df[laborRegion_df.Region == 'AlJouf']
```

```python
plt.plot(riyadh_df.Period, riyadh_df.Labor)
plt.plot(makkah_df.Period, makkah_df.Labor)
plt.plot(madinah_df.Period, madinah_df.Labor)
plt.plot(qassim_df.Period, qassim_df.Labor)
plt.plot(east_df.Period, east_df.Labor)
plt.plot(asir_df.Period, asir_df.Labor)
plt.plot(tabuk_df.Period, tabuk_df.Labor)
plt.plot(hail_df.Period, hail_df.Labor)
plt.plot(north_df.Period, north_df.Labor)
plt.plot(jazan_df.Period, jazan_df.Labor)
plt.plot(najran_df.Period, najran_df.Labor)
plt.plot(albaha_df.Period, albaha_df.Labor)
plt.plot(aljouf_df.Period, aljouf_df.Labor)

plt.xlabel("Period")
plt.ylabel("Labor Force")
plt.legend(["Riyadh", "Makkah", "Madinah", "Qassim", "East Prov.", "Asir", "Tabuk", "Hail", "North.Border", "Jazan", "Najran", "AL-Baha", "AL-Jouf"])
F = plt.gcf()
Size = F.get_size_inches()
F.set_size_inches(Size[0]*5, Size[1]*3, forward=True)
plt.show()
```
